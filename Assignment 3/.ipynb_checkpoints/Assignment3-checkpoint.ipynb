```python
### import os, struct
import matplotlib as plt
from array import array as pyarray
%matplotlib inline
from numpy import append, array, int8, uint8, zeros as np
import numpy.linalg as LA
def load_mnist(dataset="training", digits=range(10), path=''):
    
    """
    Adapted from: http://cvxopt.org/applications/svm/index.html?highlight=mnist
    """

    if dataset == "training":
        fname_img = os.path.join(path, 'train-images-idx3-ubyte')
        fname_lbl = os.path.join(path, 'train-labels-idx1-ubyte')
    elif dataset == "testing":
        fname_img = os.path.join(path, 't10k-images-idx3.idx')
        fname_lbl = os.path.join(path, 't10k-labels-idx1.idx')
    else:
        raise ValueError("dataset must be 'testing' or 'training'")

    flbl = open(fname_lbl, 'rb')
    magic_nr, size = struct.unpack(">II", flbl.read(8))
    lbl = pyarray("b", flbl.read())
    flbl.close()

    fimg = open(fname_img, 'rb')
    magic_nr, size, rows, cols = struct.unpack(">IIII", fimg.read(16))
    img = pyarray("B", fimg.read())
    fimg.close()

    ind = [ k for k in range(size) if lbl[k] in digits ]
    N = len(ind)

    images = zeros((N, rows, cols), dtype=uint8)
    labels = zeros((N, 1), dtype=int8)
    for i in range(len(ind)):
        images[i] = array(img[ ind[i]*rows*cols : (ind[i]+1)*rows*cols ]).reshape((rows, cols))
        labels[i] = lbl[ind[i]]

    return images, labels
```

```python
from pylab import *
from numpy import *
import scipy.sparse as sparse
import scipy.linalg as linalg

images, labels = load_mnist('training', digits=[0,5])

# converting from NX28X28 array into NX784 array
flatimages = list()
for i in images:
    flatimages.append(i.ravel())
X = np.asarray(flatimages)
```

```python
def writeExcelData(x,excelfile,sheetname,startrow,startcol):
    from pandas import DataFrame, ExcelWriter
    from openpyxl import load_workbook
    df=DataFrame(x)
    book = load_workbook(excelfile)
    writer = ExcelWriter(excelfile, engine='openpyxl') 
    writer.book = book
    writer.sheets = dict((ws.title, ws) for ws in book.worksheets)
    df.to_excel(writer, sheet_name=sheetname,startrow=startrow-1, startcol=startcol-1, header=False, index=False)
    writer.save()
    writer.close()
```

```python
def histClassifier(X,T, Bx ,By ,xmin,xmax):

    Hpositive = np.zeros((Bx,Bx)).astype('int32')
    Hnegative = np.zeros((By,By)).astype('int32')
    binindex1stcol = (np.round((Bx-1)*(X['1stcol'] - xmin['1stcol'])/(xmax['1stcol'] - xmin['1stcol']))).astype('int32')
    binindex2ndcol = (np.round((By-1)*(X['2ndcol'] - xmin['2ndcol'])/(xmax['2ndcol'] - xmin['2ndcol']))).astype('int32')
    for (i,b) in enumerate(binindex1stcol):
        if T[i] == 5:
            Hpositive[b][binindex2ndcol[i]]+=1
        elif T[i] == 0:
            Hnegative[b][binindex2ndcol[i]]+=1
    return [Hpositive,Hnegative]
[Hpositive,Hnegative] = histClassifier(train,train['label'],25,25,xmin,xmax)
```

```python
import math
def pdf(mean, coMatrix, coDet, x):
    var = float(coDet)**0.5
    denom = (2*math.pi*var)
    xsubmean = np.matrix(x-mean)
    invcovariance = np.linalg.inv(coMatrix)
    meanInvProduct = xsubmean* invcovariance
    xsubMeanTranspose= np.transpose([np.array(x-mean)])
    num = math.exp(-1 *meanInvProduct *xsubMeanTranspose /2 )
    #print("pdf ", num/denom)
    return num/denom

def positveBaysian(meanPositive,meanNegative,covPositive,covNegative,Positivedet,negativedet,x,y):
    pdfPositive = pdf(meanPositive[0:2],covPositive[0:2],Positivedet,[x,y])
    pdfNegative = pdf(meanNegative[0:2],covNegative[0:2],negativedet,[x,y])
    ProbPositive = npos.as_matrix()[0] * pdfPositive / (nneg.as_matrix()[0] * pdfNegative + npos.as_matrix()[0] * pdfPositive  )
    return ProbPositive

pr5baysian = positveBaysian(meanPositive,meanNegative,covPositive,covNegative,Positivedet,negativedet,pp[0],pp[1])

print("Probabilty of 5 ",pr5baysian)
```
